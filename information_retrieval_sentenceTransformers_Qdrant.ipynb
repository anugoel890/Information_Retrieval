```python
import pandas as pd
df = pd.read_csv('sample_data/wines.csv')
df = df[df['variety'].notna()] # remove any NaN values as it blows up serialization
data = df.to_dict('records')
df
```

```python
from qdrant_client import models, QdrantClient
from sentence_transformers import SentenceTransformer
```

This is used for generating embeddings from text descriptions. SentenceTransformer is a library for sentence and text embeddings, which converts textual input into fixed-size dense vectors suitable for similarity calculations.

```python
#create embedding using sentence-transformer
encoder = SentenceTransformer('all-MiniLM-L6-v2')
```

Qdrant is a vector database and client library for similarity search on vector data. It provides functionality for storing vectors, performing similarity searches, and managing collections of vector data.

```python
# create in memory instance of Qdrant vector database
qdrant = QdrantClient(":memory:")
```

```python
# Define the vector database collection parameters
qdrant.recreate_collection(
    collection_name="top_wines",
    vectors_config=models.VectorParams(
        size=encoder.get_sentence_embedding_dimension(),  # Vector size defined by used model
        distance=models.Distance.COSINE  # Specify cosine distance metric
    )
)
```

```python
# Upload wine descriptions as vectors to the Qdrant collection
qdrant.upload_records(
    collection_name="top_wines",
    records=[
        models.Record(
            id=idx,
            vector=encoder.encode(doc["notes"]).tolist(),  # Encode wine descriptions as vectors
            payload=doc
        ) for idx, doc in enumerate(data)  # Enumerate through the wine data and convert each description into a vector
    ]
)
```

```python
# Search for similar wines based on a given query vector
hits = qdrant.search(
    collection_name="top_wines",
    query_vector=encoder.encode("A wine from Mendoza Argentina").tolist(),  # Encode the query as a vector
    limit=3  # Limit the number of search results to 3
)

```

```python
# Print the top hits along with their scores
for hit in hits:
    print(hit.payload, "score:", hit.score)
```
